```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv("iris.csv")
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier(criterion='entropy')
a = df.iloc[:,:-1]
b = df.iloc[:,-1]
dtclassifier.fit(a,b)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtclassifier,filled=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier(criterion='gini')
a=df.iloc[:,:-1]
b=df.iloc[:,-1]
dtclassifier.fit(a,b)

plt.figure(figsize=(15,10))
tree.plot_tree(dtclassifier, filled = True)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier(criterion='gini', max_depth=2)
a=df.iloc[:,:-1]
b=df.iloc[:,-1]
dtclassifier.fit(a,b)

plt.figure(figsize=(15,10))
tree.plot_tree(dtclassifier, filled = True)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier(criterion='gini', max_depth=8)
a=df.iloc[:,:-1]
b=df.iloc[:,-1]
dtclassifier.fit(a,b)

plt.figure(figsize=(15,10))
tree.plot_tree(dtclassifier, filled = True)
```

```python
from sklearn.model_selection import GridSearchCV
a2 = df.iloc[:,:-1]
b2 = df.iloc[:,-1]

param_grid = {
    'criterion' : ['gini','entropy'],
    'max_depth' : [2,3,4,5],
    'min_samples_split' : [2,3,4,5],
    'min_samples_leaf' : [1,2,3,4]
}
```

```python
grid_serach = GridSearchCV(dtclassifier,param_grid,cv = 5)
grid_serach.fit(a2,b2)
```

```python
best_params = grid_serach.best_params_
best_params
```

```python
best_classifier = grid_serach.best_estimator_
best_classifier
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(best_classifier,filled=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(a2,b2,test_size=0.2)
```

```python
from sklearn import tree
DT = tree.DecisionTreeClassifier()
DT.fit(x_train,y_train)
y_pred = DT.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(f'Model accuracy score : {100*accuracy_score(y_test,y_pred):0.2f}%')
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print('confusion Matrix')
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
