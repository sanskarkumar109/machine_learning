```python
import pandas as pd
df = pd.read_csv("playTennis.csv")
df.head()
```

```python
y = df['Play Tennis']
x = df.drop('Play Tennis', axis = 1)
y
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x.Outlook = le.fit_transform(df['Outlook'])
x.Temperature = le.fit_transform(df['Temperature'])
x.Humidity = le.fit_transform(df['Humidity'])
x.Wind = le.fit_transform(df['Wind'])
x
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x,y)
predicted = model.predict([[2,0,0,1]])
print('Predicted Value',predicted)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
from sklearn.naive_bayes import GaussianNB
model_gnb = GaussianNB()
model_gnb.fit(x_train,y_train)
y_pred = model_gnb.predict(x_test)

from sklearn.metrics import accuracy_score
print(f'Model accuracy score : {100*accuracy_score(y_test,y_pred):0.2f}%')
print()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print("confusion metrix")
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
